# Quadratic Regression (Jupyter Notebook)

Este notebook muestra cómo ejecutar la regresión cuadrática incremental con los mismos componentes de `quadratic_regression.py`.

```python
from pathlib import Path

from quadratic_regression import (
    FitConfig,
    LeastSquaresQuadraticFittingStrategy,
    QuadraticRegressionService,
    read_points,
    write_snapshots,
)
```

```python
# Configuración equivalente a los parámetros por defecto del CLI
config = FitConfig(
    tolerance=0.12,
    max_failed_predictions=4,
    significant_digits=6,
    min_points=5,
)

service = QuadraticRegressionService(
    config=config,
    strategy=LeastSquaresQuadraticFittingStrategy(),
)
```

```python
input_path = Path('input.txt')
output_path = Path('output_from_notebook.txt')

points = read_points(input_path)
snapshots = service.run(points)
write_snapshots(output_path, snapshots)

print(f'Puntos procesados: {len(points)}')
print(f'Snapshots generados: {len(snapshots)}')
print(f'Salida escrita en: {output_path.resolve()}')
```

```python
# Vista rápida de las primeras filas de salida
for line in output_path.read_text(encoding='utf-8').splitlines()[:10]:
    print(line)
```
